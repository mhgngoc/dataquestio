```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

women_degrees = pd.read_csv("percent-bachelors-degrees-women-usa.csv")
majors = list(women_degrees.columns)[1:]
```

```python
stem_cats = ['Engineering', 'Computer Science', 'Psychology','Biology', 'Physical Sciences', 'Math and Statistics']
```

```python
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
```

```python
fig = plt.figure(figsize=(6,10))
for i in range(6):
    ax = fig.add_subplot(3,2,i+1)
    ax.plot(women_degrees["Year"],women_degrees[stem_cats[i]],
    c=cb_dark_blue,label="Women",linewidth=3)
    ax.plot(women_degrees["Year"],100-women_degrees[stem_cats[i]],
    c=cb_orange,label="Men",linewidth=3)
    ax.tick_params(bottom="off",top="off",left="off",right="off")
    ax.set_ylim(0,100)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_title(stem_cats[i])
plt.show()
```

```python
len(majors)
fig = plt.figure(figsize=(9,20))
for i in range(17):
    ax = fig.add_subplot(6,3,i+1)
    ax.plot(women_degrees["Year"],women_degrees[majors[i]],
    c=cb_dark_blue,label="Women",linewidth=3)
    ax.plot(women_degrees["Year"],100-women_degrees[majors[i]],
    c=cb_orange,label="Men",linewidth=3)
    ax.tick_params(bottom="off",top="off",left="off",right="off",
    labelbottom="off")
    ax.set_ylim(0,100)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_title(majors[i])
    if i==14 or i==15 or i==16:
        ax.tick_params(labelbottom="on")
    if i==16:
        ax.legend(loc="upper right")
plt.show()
```
